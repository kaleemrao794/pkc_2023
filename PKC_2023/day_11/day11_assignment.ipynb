## **2. Visualization on Iris Dataset**

**Importing the libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

**Import dataset**

```python
phool = sns.load_dataset('iris')
phool.info()
```

```python
phool.isnull().sum()
```

### **1-Bar-Plot**

```python
phool.species.value_counts().plot(kind='bar')
```

```python
sns.barplot(x='species', y='sepal_width', data=phool)
```

### **2-Scatter-Plot**

```python
X=phool[['sepal_length', 'petal_length']]
Y=phool['species'] == 'virginica'      # This can be changed to 'setosa' and 'versicolor' (unique values)
plt.scatter(X['sepal_length'], X['petal_length'], c=Y)
                                       # yellow = selected specie, blue = other species (as per the above line)
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,5))
species_list=['setosa', 'versicolor', 'virginica']

for i in range(3):
    subset_phool = phool[phool['species'] == species_list[i]]
    ax[i].scatter(subset_phool['sepal_width'], subset_phool['petal_width'],
                  c=subset_phool['species'].map({'setosa':'r', 'versicolor':'y', 'virginica':'b'}))
    ax[i].set_title('species = ' + str(species_list[i]))
    ax[i].set_xlabel('sepal_width')
    ax[i].set_ylabel('petal_width')
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=phool, palette='Set1')
```

### **3-Hist-Plot**

```python
phool.hist(figsize=(10,10), grid=False, layout=(2,2),
            bins=20, color='r', edgecolor='k', alpha=0.7)
```

### **4-Pair-Plot**

```python
sns.pairplot(phool, diag_kind='kde', hue='species')
```

### **5-Box-Plot**

```python
plt.figure(figsize=(5,5))
sns.boxplot(x='petal_length', data=phool)
plt.show()
```

### **6-Boxen-Plot**

```python
plt.figure(figsize=(5,5))
sns.boxenplot(x='sepal_width', data=phool)
plt.show()
```

### **7-Sub-Plots**

```python
# Create a figure and subplots for each feature
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axes = axes.flatten()

# Create a list of features to iterate over
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Iterate over the features and plot the data for each species on its corresponding subplot
for ax, feature in zip(axes, features):
    sns.boxplot(x='species', y=feature, data=phool, ax=ax)
    ax.set_title(f'{feature} of different Iris species')
    ax.set_xlabel(feature)
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

### **8-Strip-Plot**

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axes = axes.flatten()
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for ax, feature in zip(axes, features):
    sns.stripplot(x='species', y=feature, data=phool, ax=ax)
    ax.set_title(f'{feature} of different Iris species')
    ax.set_xlabel(feature)
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

### **9-Heatmap**

```python
corr = phool.corr()
plt.figure(figsize=(8,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

### **10-Violin-Plot**

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axes = axes.flatten()
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for ax, feature in zip(axes, features):
    sns.violinplot(x='species', y=feature, data=phool, ax=ax)
    ax.set_title(f'{feature} of different Iris species')
    ax.set_xlabel(feature)
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

### **Step-8: Selecct a Column by dtype**

```python
kashti.dtypes
```

```python
# Only select those having numeric types
kashti.select_dtypes(include=['number']).head()
```

```python
# Only select those having object types
kashti.select_dtypes(include=['object']).head()
```

```python
# Only select those having multiple types
kashti.select_dtypes(include=['object', 'category']).head()
```

```python
# use of exclude function
kashti.select_dtypes(exclude=['object']).head()
```

### **Step-9: Convert String into number**

```python
df = pd.DataFrame({'Col_A':['1','2','3','4','5','6','7','8','9'],
                  'Col_B':['10','11','12','13','14','15','16','17','18']})
df
```

```python
df.dtypes
```

```python
# type conversion
df = df.astype({'Col_A': 'float64', 'Col_B' : 'int64'})
df.dtypes
```
